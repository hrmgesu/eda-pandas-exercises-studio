### 1. Read data

```python
# import pandas
import pandas as pd

# create your data variable using .read_csv 
data = pd.read_csv('./IMDB-Movie-Data.csv')

# create your data_indexed variable here
# Hint: use the index_col parameter to specify which column should be the index (try 'Title')
data_indexed = pd.read_csv('./IMDB-Movie-Data.csv', index_col="Title")

# Read data with specified explicit index
# This allows you to use movie titles as row labels instead of numeric indexes

print(data_indexed)
```

### 4. Data Selection -- Indexing and Slicing

```python
#  Extract data as series
# define and print the variable 'genre'
genre = data['Genre']
print(genre)
```

```python
# Extract the data as a dataframe and print it
genre = data[['Genre']]
print(genre)

```

```python
# Extract data using rows
# loc and iloc are two functions that can be used to slice data from specific row indexes.

# loc function
# We will use the loc function to index and slice that dataframe using the specified 'Title'.
data_indexed.loc[['Suicide Squad']][['Genre','Actors','Director','Rating','Revenue (Millions)']]
```

```python
# iloc function
# Here, iloc is used to slice data using integer indexes.
data.iloc[10:15][['Title','Rating','Revenue (Millions)']]
```

### 5. Data Selection - Based on Conditional Filtering

```python
# Use this code block to select movies that have been released between 2010-2016, with a rating of less than 6.0, but generated significant revenue.
# For "significant revenue", filter for movies in the top 25% of revenue (use data['Revenue (Millions)'].quantile(0.75))
data[((data['Year'] >= 2010) & (data['Year'] <= 2016))
      & (data['Rating'] < 6.0)
      & (data['Revenue (Millions)'] > data['Revenue (Millions)'].quantile(0.95))]
```
